```python
def having_tick(x):
    if len(x)>3:
        x1,x2 = x.split('-')
        return str((int(x1)+int(x2))/2).split('.')[0]
    else:
        return x
```

```python
punjab['Age Bracket']  = punjab["Age Bracket"].apply(having_tick)
```
